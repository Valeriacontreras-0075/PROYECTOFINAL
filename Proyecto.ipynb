```python
import pandas as pd
import matplotlib.pyplot as plt

# Paleta pastel personalizada
pastel_colors = ["#C8A2C8",  # lila pastel
                 "#AEC6CF",  # azul pastel
                 "#FFFACD",  # amarillo pastel
                 "#FFB6C1",  # rosa pastel
                 "#77DD77"]  # verde pastel

# Cargar archivos CSV (asegrate de subirlos en Colab)
clientes = pd.read_csv("Clientes.csv")
eventos = pd.read_csv("Eventos.csv")
inventario = pd.read_csv("Inventario.csv")
pedidos = pd.read_csv("Pedidos.csv")
productos = pd.read_csv("Productos.csv")

# 1. Clientes por tipo (grfico circular)
plt.figure(figsize=(6,6))
clientes["TipoCliente"].value_counts().plot(
    kind="pie", autopct="%1.1f%%", colors=pastel_colors, startangle=90
)
plt.title("Clientes por Tipo", fontsize=14)
plt.ylabel("")
plt.show()

# 2. Eventos por tipo (grfico circular)
plt.figure(figsize=(6,6))
eventos["tipoEvento"].value_counts().plot(
    kind="pie", autopct="%1.1f%%", colors=pastel_colors, startangle=90
)
plt.title("Eventos por Tipo", fontsize=14)
plt.ylabel("")
plt.show()

# 3. Inventario por proveedor (grfico de barras)
plt.figure(figsize=(6,4))
inventario.groupby("proveedor")["stockActual"].sum().plot(
    kind="bar", color=pastel_colors[1]
)
plt.title("Inventario Actual por Proveedor")
plt.ylabel("Unidades")
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

# 4. Pedidos por tipo y mtodo de pago (grfico de barras apiladas)
plt.figure(figsize=(6,4))
pedidos.groupby(["tipoPedido", "metodoPago"]).size().unstack().plot(
    kind="bar", stacked=True, color=pastel_colors[:3]
)
plt.title("Pedidos por Tipo y Mtodo de Pago")
plt.ylabel("Cantidad")
plt.xticks(rotation=0)
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

# 5. Productos por categora (grfico circular)
plt.figure(figsize=(6,6))
productos["categoria"].value_counts().plot(
    kind="pie", autopct="%1.1f%%", colors=pastel_colors, startangle=90
)
plt.title("Productos por Categora", fontsize=14)
plt.ylabel("")
plt.show()


```
